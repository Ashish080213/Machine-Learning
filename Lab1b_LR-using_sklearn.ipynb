```python
# LINEAR REGRESSION WITH ONE VARIABLE USING SKLEARN
```

```python
# ----> importing all modules <----

# imported numpy to perform calculations
import numpy as np

# imported matplotlib to visualize the data and regression LinearRegression 
import matplotlib.pyplot as plt

# imported pandas to import the data set -> csv file 
import pandas as pd

# LinearRegression is a python class which is used to perform the linear regression which is imported from sklearn
from sklearn.linear_model import LinearRegression
```

```python
# loading the csv file dataset
data = pd.read_csv('ex1data1.txt')

# ----> Separating the data into independent and dependent variables <----
# extracting the values from 1st column or index "0"
# Converting dataset into a numpy array 
# we are reshping by giving "-1" as 1st argument -> this indicates to calculate the dimension of the rows by the model automatically
# "1" as 2nd argument -> this indicates column dimension is 1
x = data.iloc[:, 0].values.reshape(-1, 1) # independent variable values
# extracting the values from 2nd column or index "0"
# Converting dataseet into a numpy array 
y = data.iloc[:, 1].values.reshape(-1, 1) # dependent variable values

# Creating a regression object
LR = LinearRegression()

# fit() --> to perform linear regression we need to call fit method (fit training data)
LR.fit(x, y)

# y_prediction consist of prediction values
# by passign x values to the function predict() --> returns the learned label for each object in the array
y_prediction = LR.predict(x)
```

```python

# plt is used to plot
# scatter and plot are the classes/methods in mathplotlib module which is used to visualize the data set and the regression line
plt.scatter(x, y)
plt.plot(x, y_prediction, color='red')

# show() method displays the figures
plt.show()
```
