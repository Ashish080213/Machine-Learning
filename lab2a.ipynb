```python
# imported numpy to perform calculations
import numpy as np
# imported pandas to import the data set -> csv file 
import pandas as pd
```

```python
# loading the iris dataset csv file
# skiprows is used to skip the 1st row of the dataset
data = pd.read_csv('iris_dataset.csv', skiprows=1,header=None)
data
```

```python
# Split the dataset into input features (x)

# extracting the values from 1st column or index "0"
# Converting dataset into a numpy array 

x = data.iloc[:,:-1].values
x
```

```python
# Split the dataset into target variable (y)

# extracting the values from 1st column or index "0"
# Converting dataset into a numpy array 

y1 = data.iloc[:,-1].values
y1
```

```python
# we are reshping by giving "-1" as 1st argument -> this indicates to calculate the dimension of the rows by the model automatically
# "1" as 2nd argument -> this indicates column dimension is 1

y = y1.reshape(-1,1)
y
```

```python
from sklearn.model_selection import train_test_split
# Splitting the dataset into training and test sets
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=41)

# we will have different classifiers in sklearn.tree
# we have imported tree from sklearn to get decision tree classifier
from sklearn import tree

# Fit a decision tree classifier to the training set
# max_depth = 3 in order to limit upto three layers
# "entropy" is the method of approach to find Decision value
classifier = tree.DecisionTreeClassifier(min_samples_split=3,max_depth=3,criterion="entropy")

# fit is used to train the model
classifier.fit(xtrain,ytrain)

# score is used to find the accuracy of the classifier on the test set
score = classifier.score(xtest,ytest)
score
```

```python
# predict is used to make predictions on the test set to get predicted y values
y_pred = classifier.predict(xtest)
y_pred
```

```python
y_pred1 = y_pred.reshape(-1,1)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
# accuracy_score is imported to show the accuracy of the predictions
accuracy = accuracy_score(ytest, y_pred)
print(accuracy)
```

```python
# imported matplotlib to visualize the data
import matplotlib.pyplot as plt
# plot_tree(classifier, filled=True) is used to plot the decision tree with filled nodes
tree.plot_tree(classifier,max_depth=3,filled=True)
# plt.show() to display the plot
plt.show()
```
